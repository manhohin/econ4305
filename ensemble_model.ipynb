```python
import json
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn import metrics
```

# MLP

```python
with open('MLP_y_pred_list.json', 'r') as f:
    MLP_y_pred_list = json.load(f)
    MLP_y_pred_list = [item for sublist in MLP_y_pred_list for item in sublist]
with open('MLP_y_test_list.json', 'r') as f:
    MLP_y_test_list = json.load(f)

# for the testset
ConfusionMatrixDisplay.from_predictions(MLP_y_test_list[9:], MLP_y_pred_list[9:])
print(classification_report(MLP_y_test_list[9:], MLP_y_pred_list[9:]))
```

```python
MLP_f1 = [0.66, 0.66]
```

# LSTM

```python
with open('LSTM_y_pred_list.json', 'r') as f:
    LSTM_y_pred_list = json.load(f)
    LSTM_y_pred_list = [item for sublist in LSTM_y_pred_list for item in sublist]
with open('LSTM_y_test_list.json', 'r') as f:
    LSTM_y_test_list = json.load(f)
    
# for the testset
ConfusionMatrixDisplay.from_predictions(LSTM_y_test_list, LSTM_y_pred_list)
print(classification_report(LSTM_y_test_list, LSTM_y_pred_list))
```

```python
LSTM_f1 = [0.72, 0.78]
```

# Random Forest

```python
with open('Random_Forest_y_pred_list.json', 'r') as f:
    Random_Forest_y_pred_list = json.load(f)
with open('Random_Forest_y_test_list.json', 'r') as f:
    Random_Forest_y_test_list = json.load(f)

Random_Forest_y_pred_val = Random_Forest_y_pred_list[:-100]
Random_Forest_y_val = Random_Forest_y_test_list[:-100]
Random_Forest_test_pred = Random_Forest_y_pred_list[-100:]
Random_Forest_test_y = Random_Forest_y_test_list[-100:]

# for val set
ConfusionMatrixDisplay.from_predictions(Random_Forest_y_val, Random_Forest_y_pred_val)
print(classification_report(Random_Forest_y_val, Random_Forest_y_pred_val))

# for test set
ConfusionMatrixDisplay.from_predictions(MLP_y_test_list[9:], Random_Forest_test_pred[9:])
print(classification_report(MLP_y_test_list[9:], Random_Forest_test_pred[9:]))
```

```python
Random_Forest_f1 = [0.74, 0.81]
```

# XGboost

```python
with open('XGboost_y_pred_list.json', 'r') as f:
    XGboost_y_pred_list = json.load(f)
with open('XGboost_y_test_list.json', 'r') as f:
    XGboost_y_test_list = json.load(f)

XGboost_val_Y = XGboost_y_test_list[:-100]
XGboost_test_Y = XGboost_y_test_list[-100:]
XGboost_val_pred = XGboost_y_pred_list[:-100]
XGboost_test_pred = XGboost_y_pred_list[-100:]

# For the validation set
ConfusionMatrixDisplay.from_predictions(XGboost_val_Y, XGboost_val_pred)
print(classification_report(XGboost_val_Y, XGboost_val_pred))

# For the test set 
ConfusionMatrixDisplay.from_predictions(MLP_y_test_list[9:], XGboost_test_pred[9:])
print(classification_report(MLP_y_test_list[9:], XGboost_test_pred[9:]))
```

```python
XGboost_f1 = [0.68, 0.78]
```
